```python
import plotly.graph_objects as go
import sqlalchemy as db
import pandas as pd
import plotly.express as px
import os
from dotenv import load_dotenv
load_dotenv()

postgres_username = os.getenv('POSTGRES_USERNAME')
postgres_password = os.getenv('POSTGRES_PASSWORD')
postgres_host = os.getenv('POSTGRES_HOST')

eng = db.create_engine(f'postgresql://{postgres_username}:{postgres_password}@{postgres_host}/Recreation')

sql = "SELECT * FROM rec_acts WHERE act_type <> 'AlpineSki'"

df = pd.read_sql_query(sql, eng).sort_values(by=['act_date'], ascending=False)
```

```python
df['hour'] = df['start_time'].astype('str').str.split(':').str[0].astype('int')
df['year'] = df['act_date'].astype('str').str.split('-').str[0].astype('int')
df['month'] = df['act_date'].astype('str').str.split('-').str[1].astype('int')
df['day'] = df['act_date'].astype('str').str.split('-').str[2].astype('int')
df['duration_hours'] = (df['duration'] / 60 / 60).round(1)
df['duration_minutes'] = (df['duration'] / 60).round(1)
df
```

```python
year = df.groupby('year').agg({'duration_hours': 'sum'}).reset_index()
month = df.groupby('month').agg({'duration_hours': 'sum'}).reset_index()
month
```
